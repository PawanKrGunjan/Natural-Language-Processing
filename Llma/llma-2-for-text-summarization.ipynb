# Load the model

```python
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, AutoModelForCausalLM
import torch

# Load LLAMA model and tokenizer
model_name = "zaanind/llma-2-7b-tuned"
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

# Llma for Text Generations

```python
from transformers import AutoModelForCausalLM

# Load the model
model = AutoModelForCausalLM.from_pretrained(model_name)

# Prompt for text generation
prompt_text = "AI is,"

# Tokenize the prompt
input_ids = tokenizer.encode(prompt_text, return_tensors="pt")

# Generate text
output = model.generate(input_ids, max_length=100, num_return_sequences=1, temperature=0.9)

# Decode and print the generated text
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
print("Generated Text:", generated_text)
```
