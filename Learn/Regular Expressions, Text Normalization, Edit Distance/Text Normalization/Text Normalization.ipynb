## Tokenization

```python
text = 'That U.S.A. poster-print costs $80.5 i.e $81, out of 100 it is 80.5% i.e 81% ...'

patterns = "([A-Z]+\.)+|\$?\d+(\.\d+)?%?|\w+(-\w+)?(\.\w)?"

for i in re.finditer(patterns,text):
    print(i.group())
```

```python
patterns = "([A-Z]+\.)+|\$?\d+(\.\d+)?%?|\w+(-\w+)?(\.\w)?"
Token = []
for i in re.finditer(patterns, Txt):
    Token.append(i.group())
    
print('There are total',len(Token),'words in the corpus')

## Unique Token
Tokens = list(set(Token))
print('There are only',len(Tokens),'unique tokens in the corpus')
```

```python
# Python code to find frequency of each word
def frequency(Text):
    
    # break the string into list of words
    patterns = "([A-Z]+\.)+|\$?\d+(\.\d+)?%?|\w+(-\w+)?(\.\w)?"
    Token = []
    for i in re.finditer(patterns, Text.lower()):
        Token.append(i.group())
    Tokens = list(set(Token))     
    frequency = {}      
    for token in Tokens:
        frequency[token] =Token.count(token) 
    return frequency

freq = frequency(Txt)
```
