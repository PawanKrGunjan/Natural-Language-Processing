```python
import json
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
import wget
url='https://storage.googleapis.com/laurencemoroney-blog.appspot.com/sarcasm.json'
filename = wget.download(url)
filename
```

```python
data=pd.read_json(filename)
data.head()
```

```python
data.info()
```

```python
df = data.iloc[:,1:]
df.head()
```

```python
df.shape
```

```python
sentence = df.headline
labels = df.is_sarcastic
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train,y_test = train_test_split(sentence, labels, train_size=0.80,random_state=2)
```

```python
X_train.shape, X_test.shape,y_train.shape,y_test.shape
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(X_train)
```

```python
word_index = tokenizer.word_index
len(word_index)
```

```python
training_sequences = tokenizer.texts_to_sequences(X_train)
training_sequences[0]
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

training_padded = pad_sequences(training_sequences, 
                                maxlen=100, 
                                padding='post', 
                                truncating='post')
training_padded[0]
```

```python
testing_sequences = tokenizer.texts_to_sequences(X_test)

testing_padded = pad_sequences(testing_sequences, 
                               maxlen=100, 
                               padding='post', 
                               truncating='post')
```

```python
testing_padded[0]
```

```python
#Layer Normalization
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(10000, 16, input_length=100),
    tf.keras.layers.GlobalAveragePooling1D(),
    tf.keras.layers.Dense(24, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(training_padded, 
                    y_train,
                    epochs=10, 
                    validation_data=(testing_padded, y_test), 
                    verbose=2)
```

```python
pd.DataFrame(history.history).head()
```

### Plot accuracy vs val_accuracy & loss vs  val_loss curve

```python
import matplotlib.pyplot as plt

def plot_graphs(history, string):
    plt.plot(history.history[string])
    plt.plot(history.history['val_'+string])
    plt.xlabel("Epochs")
    plt.ylabel(string)
    plt.legend([string, 'val_'+string])
    plt.title(f"{string} vs {'val_'+string}", fontsize = 15)
    plt.show()

plot_graphs(history, "accuracy")
plot_graphs(history, "loss")
```

```python
model.evaluate(testing_padded,y_test)
```

```python
pred_labels= [1 * (i[0]>=0.5) for i in pred]
pred_labels[:5]
```

```python
# Reverse the word_index
reverse_word_index = dict([(value, key) for (key, value) in word_index.items()])
reverse_word_index[1084]
```

```python
# Decode the sentence
def decode_sentence(text):
    word=list()
    for i in text:
        if i != 0:
            word.append(reverse_word_index[i])
    sentence = ' '.join(word)
    return 'sentence :',sentence
```

```python
decode_sentence(testing_padded[0])
```

```python
e = model.layers[0]
e
```

```python
weights = e.get_weights()[0]
weights[0]
```

```python
vocab_size, embedding_dim = weights.shape
vocab_size
```

# Save tokenize set & Predicted value

```python
import io

out_v = io.open('vecs.tsv', 'w', encoding='utf-8')
out_m = io.open('meta.tsv', 'w', encoding='utf-8')
for word_num in range(1, vocab_size):
    word = reverse_word_index[word_num]
    embeddings = weights[word_num]
    out_m.write(word + "\n")
    out_v.write('\t'.join([str(x) for x in embeddings]) + "\n")
out_v.close()
out_m.close()
```
